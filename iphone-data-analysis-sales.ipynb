It returns a new DataFrame containing the following statistical metrics for each numeric column:

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

What df.isnull().sum() Returns
When you call df.isnull().sum(), it performs the following operations:

df.isnull(): Creates a DataFrame of the same shape as df but with boolean values (True for missing values and False for non-missing values).

.sum(): Aggregates the boolean values for each column, summing True values (which are considered as 1) and providing the total count of missing values per column.

```python
highest_rated_df=df.sort_values(by=["Star Rating"],ascending=False)
highest_rated_df=highest_rated_df.head(10)
#will return first 10 rows of which iphones are highest rated rationgs like: 4.7, 4.5 etc.
print(highest_rated_df)
```

```python
label=highest_rated_df['Product Name'].value_counts().index
print(label)
```

value_counts(): Counts the occurrences of each unique value in the 'Product Name' column.
.index: Gets the index of the resulting Series, which are the unique product names sorted by their frequency.

```python
counts=highest_rated_df["Number Of Ratings"]
print(counts)
```

```python
figure=px.bar(highest_rated_df,x=label,y=counts,title="Number of Ratings of Highest Rated Iphones")
figure.show()
```

```python
label2=highest_rated_df['Product Name']
print(label2)
```

```python
figure2=px.bar(highest_rated_df,x=label2,y=counts,title="Alternative Method")
figure2.show()
```

```python
#lets do the same 
highly_reviewed_df=df.sort_values(by=["Number Of Reviews"],ascending=False)
print(highly_reviewed_df.head(10))
highly_reviewed_df=highly_reviewed_df.head(10)
```

```python
label3=highly_reviewed_df["Product Name"]
count3=highly_reviewed_df["Number Of Reviews"]
figure3=px.bar(highly_reviewed_df,x=label3,y=count3,title="Top 10 Highly Reviewed Iphones")
figure3.show()
```

```python
#lets find connection between number of reviews and discount
#we already have df sorted by highly reviewed 
#count3 is series of number of reviews
label4=highly_reviewed_df["Discount Percentage"]
figure4=px.bar(highly_reviewed_df,x=label4,y=count3,title="Relation Between Disount and Reviews")
figure4.show()
```

```python
label4=highly_reviewed_df["Sale Price"]
figure4=px.bar(highly_reviewed_df,x=label4,y=count3,title="Relation Between Sale Price and Reviews")
figure4.show()
```
